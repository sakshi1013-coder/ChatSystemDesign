1: UserShardManager Code
Here routing logic is changed so that it is based on user_id instead of channel_id

```python
class Shard:
    def __init__(self, shard_id):
        self.id = shard_id
        self.messages = []

    def store(self, message):
        self.messages.append(message)


class ShardManager:
    def __init__(self, num_shards):
        self.num_shards = num_shards
        self.shards = [Shard(i) for i in range(num_shards)]

    def stats(self):
        for shard in self.shards:
            print(f"Shard {shard.id}: {len(shard.messages)} messages")
```

```python
class UserShardManager(ShardManager):

    def get_shard(self, user_id):
        return self.shards[user_id % len(self.shards)]

    def send_message(self, message):
        shard = self.get_shard(message.user_id)
        shard.store(message)
```

2: Message Class
Here, it is defines the structure of a message
Each message contains:
user_id → who sent it
channel_id → where it was sent
content → actual message

```python
class Message:
    def __init__(self, user_id, channel_id, content):
        self.user_id = user_id
        self.channel_id = channel_id
        self.content = content
```

3: Simulation
Here, it is creating 3 shards
Sends messages from:
Many normal users
One heavy user (influencer)

```python
manager = UserShardManager(3)

# Normal users
for i in range(100):
    msg = Message(user_id=i, channel_id=1, content="Hello")
    manager.send_message(msg)

# One influencer (heavy load)
for _ in range(5000):
    msg = Message(user_id=999, channel_id=1, content="Influencer spam")
    manager.send_message(msg)

manager.stats()
```
