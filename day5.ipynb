Here we have 3 machines but the data is not assigned. Most of the time it is assigned randomly or using global which can be hard to retrieve later. So we need to choose a simple routing rule conceptual based

1. Shard Class
this represents one machine or shard. and has each one id which is unique identifier and a message that is its own storage.

Before we had everthing on server and now the data is splitted across the machines or shards

```python
class Shard:
    def __init__(self, shard_id):
        self.id = shard_id
        self.messages = []

    def store(self, message):
        self.messages.append(message)
```

2. ShardManager
Here, this is creating multiple shards or machines and stores them in a list. get_shard, decides where which shard will store the message. Each shard represents a separate machine with its own storage. This step sets up the infrastructure for distributing data, but does not yet define how messages are routed between shards.

```python
class ShardManager:
    def __init__(self, num_shards):
        self.num_shards = num_shards
        self.shards = [Shard(i) for i in range(num_shards)]

    def get_shard(self, message):
        # Simple routing rule (not hardcoded to specific shard)
        shard_index = message.channel_id % self.num_shards
        return self.shards[shard_index]

    def send_message(self, message):
        shard = self.get_shard(message)
        shard.store(message)

    def stats(self):
        for shard in self.shards:
            print(f"Shard {shard.id}: {len(shard.messages)} messages")


```

3. Test 
This cell simulates sending multiple messages to the system and demonstrates how messages are distributed across different shards using the routing logic. It helps observe whether the load is balanced or uneven.

```python
class Message:
    def __init__(self, user_id, channel_id, content):
        self.user_id = user_id
        self.channel_id = channel_id
        self.content = content


manager = ShardManager(6)

# simulate messages
for i in range(100):
    msg = Message(user_id=i, channel_id=i % 5, content="Hello")
    manager.send_message(msg)

manager.stats()
```
