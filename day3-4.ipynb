Naive system 

```python
class Message:
    def __init__(self, user_id, channel_id, content):
        self.user_id = user_id
        self.channel_id = channel_id
        self.content = content


class ChatServer:
    def __init__(self):
        self.messages = []
    
    def send_message(self, message):
        self.messages.append(message)

    def stats(self):
        print("Total messages:", len(self.messages))
```

Test with 10 user

```python
import time

server = ChatServer()

start = time.time()

for i in range(10):
    msg = Message(i, 1, "Hello")
    server.send_message(msg)

end = time.time()

server.stats()
print("Time taken:", end - start)
```

Test with 10000 users

```python
server = ChatServer()

start = time.time()

for i in range(10000):
    msg = Message(i, 1, "Hello")
    server.send_message(msg)

end = time.time()

server.stats()
print("Time taken:", end - start)
```

Improved storage

```python
class MessageStore:
    def __init__(self, limit=5000):
        self.messages = []
        self.limit = limit

    def save_message(self, message):
        if len(self.messages) >= self.limit:
            self.messages.pop(0)  # remove oldest message

        self.messages.append(message)
```

Improved chat server

```python
class ImprovedChatServer:
    def __init__(self, store):
        self.store = store

    def send_message(self, message):
        self.store.save_message(message)

    def stats(self):
        print("Total messages:", len(self.store.messages))
```

Test Improved System for 10000

```python
server = ImprovedChatServer(MessageStore(limit=5000))

start = time.time()

for i in range(10000):
    msg = Message(i, 1, "Hello")
    server.send_message(msg)

end = time.time()

server.stats()
print("Time taken:", end - start)
```

Show memory Difference

```python
import sys

print("Memory used (approx):", sys.getsizeof(server.store.messages))
```
